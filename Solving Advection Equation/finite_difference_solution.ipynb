```python
from __future__ import division 
import numpy as np
import matplotlib.pyplot as plt

#FALSE : USEFULL
```

```python
a = 1 
t_max = 2.4
x_max = 3 
```

```python
def u_0(x):
    if (x <= 0.5 and x >= -0.5):
        return (np.cos((np.pi)*x))**2
    else:
        return 0
```

# FTBS

```python
X1 = np.arange(-3, x_max+0.1, 0.1) 
T1 = np.arange(0, t_max+(0.8*0.1), (0.8*0.1))
U1 = np.zeros((len(T1),len(X1))) 

for j in range(len(X1)):
    U1[0, j] = u_0(X1[j])

for k in range(0, len(T1)-1): 
    for j in range(1, len(X1)):
        U1[k + 1, j] = U1[k, j] + 0.8 * (U1[k, j-1] - U1[k, j])

plt.figure()
plt.plot(X1, U1[len(T1)-1,:])
plt.xlabel('x')
plt.ylabel('u(t, x): h = 1/10') 
plt.grid()
plt.title("Initial value of u: u(0, x), h = 1/10", fontsize=14)

x = (abs(U1) >= 5)
any(True in sub for sub in x) 
```

# Leap Frog

```python
X1 = np.arange(-3, x_max+0.1, 0.1) 
T1 = np.arange(0, t_max+(0.8*0.1), (0.8*0.1))
U1 = np.zeros((len(T1),len(X1))) 

for j in range(len(X1)):
    U1[0, j] = u_0(X1[j])
    
for j in range(len(X1)-1):
    U1[1, j]  = (U1[0, j] + 0.5 * 0.8 * (U1[0, j+1] - U1[0, j - 1]))
#            U1[k+1 , j] = (U1[k, j] + 0.5 * 0.8 * (U1[k, j+1] - U1[k, j - 1]))
U1[1,len(X1)-1] = U1[1,len(X1)-2] 

for k in range(1, len(T1)-1): 
    for j in range(0, len(X1)-1):
        U1[k+1 , j] = U1[k-1 , j] - 0.8* (U1[k , j+1] - U1[k , j-1])
        U1[2:,len(X1)-1] = U1[2:,len(X1)-2]

plt.figure()
plt.plot(X1, U1[len(T1)-1,:])
plt.xlabel('x')
plt.ylabel('u(t, x): h = 1/10') 
plt.grid()
plt.title("Initial value of u: u(0, x), h = 1/10", fontsize=14)

x = (abs(U1) >= 5)
any(True in sub for sub in x) 
```
