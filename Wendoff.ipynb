```python
from __future__ import division 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def WB(h,k):

    X = np.arange(0, 1+h, h) 
    T = np.arange(0, 1+k, k)
    U = np.zeros((len(X),len(T))) 
    
    for j in range(len(T)):
        U[0, j] = 0

    for i in range(len(X)):
        U[i, 0] = X[i] * X[i]  
        
    for m in range(0, len(X)-1): 
        for n in range(0, len(T)-1): 
            b = T[n]/(2+T[n])
            U[m+1, n+1]  = U[m, n] - (b)* U[m+1, n] + (b) * U[m, n+1]
            
    return U,T,X
```

# ANALYTICAL SOLUTION

```python
def Analytic(h,k):
    X = np.arange(0, 1+h, h) 
    T = np.arange(0, 1+k, k)
    U = np.zeros((len(X),len(T))) 
    
#     for j in range(len(T)):
#         U[0, j] = 0

#     for i in range(len(X)):
#         U[i, 0] = X[i] * X[i]

    for m in range(0, len(X)): 
        for n in range(0, len(T)): 
    #         b = T[n]/(2+T[n])
            if ( X[m] >= (3*T[n]*T[n]) / 2) :
                U[m, n]  = (X[m] - (3* T[n] * T[n] )/ 2) ** 2

            else :
                U[m,n] = 0
                
    return U,T,X
```
