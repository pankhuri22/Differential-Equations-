```python
import numpy as np
import matplotlib.pyplot as plt
import math 

def y1(t,y):
    return ((t**2-y**2)* np.sin(y))

h = 0.05
t0 = 0.0
y0 = -1.0

t = np.arange(0.0, 2.0+0.001, h)
y = np.zeros(t.size)
y[0] = y0

for i in range(1, t.size):
    y_intermediate = y[i-1] + h*y1(t[i-1],y[i-1])

    y[i] = y[i-1] + (h/2.0)*(y1(t[i-1],y[i-1]) + y1(t[i],y_intermediate))
    
    
plt.plot(t,y,'r-',linestyle='--')
print(y[np.where(t==0.5)])
print(y[np.where(t==1.0)])
print(y[np.where(t==1.5)])
print(y[np.where(t==2.0)])

plt.plot(0.5,y[10], 'g*')
plt.plot(1.0,y[20], 'g*')
plt.plot(1.5,y[30], 'g*')
plt.plot(2.0,y[40], 'g*')
    
plt.plot(t,y,'r-',linestyle='--')
plt.xlabel('time, h=0.05')
plt.ylabel('Y values')
plt.grid(True)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import math 

# def fixed_point(h, t, y_old,F):
#     y_0 = y_old
#     y_new = y_old
#     while(True):
#         y_new = y_0 + h * (F(t, y_old))
# #         y[i + 1] = y[i] + h*(9.0 * f(t[i+1], y[i + 1]) + 19.0 * f(t[i],y[i]) - 5.0 * f(t[i-1],y[i-1]) + f(t[i-2],y[i-2]))/24.0

#         if (abs(y_new - y_old) < 0.001):
#             break
#         y_old = y_new 
#     return y_new

def heun(h,t0,y0,f):
    t = np.arange(0.0, 2.0+0.001, h)
    y = np.zeros(t.size)
    y[0] = y0

    for i in range(1, t.size):
        y_intermediate = y[i-1] + h*f(t[i-1],y[i-1])
        y[i] = y[i-1] + (h/2.0)*(f(t[i-1],y[i-1]) + f(t[i],y_intermediate))
    
    return t,y

def backward_euler(h,t0,y0,f):
    t = np.arange(0.0, 2.0+0.001, h)
    y = np.zeros(t.size)
    y[0] = y0
    for i in range(1, t.size):
        y[i] = fixed_point(h,t[i],y[i-1],f)
#         print (temp)
    
    return t,y


# print(y.size)

def rk4(h,t0,y0,f):
    t = np.arange(0.0, 2.0+0.001, h)
    y = np.zeros(t.size)
    u_new = np.zeros(t.size)
    y[0] = y0

    for k in range(0, t.size-1):
            dt = h
            dt2 = dt/2.0
            K1 = h*f(t[k], y[k])
            K2 = h*f(t[k] + dt2, y[k] + K1*0.5)
            K3 = h*f(t[k] + dt2, y[k] + K2*0.5)
            K4 = h*f(t[k] + dt, y[k] + K3)
            y[k+1] = y[k] + (1/6.0)*(K1 + 2*K2 + 2*K3 + K4)

    return t,y


def forward_euler(h,t0,y0,f):
    t = np.arange(0.0, 2.0+0.001, h)
    y = np.zeros(t.size)
    y[0] = y0

    for i in range(1, t.size):
        y[i] = y[i-1] + f(t[i-1],y[i-1])*h
#         print(y1(t[i-1],y[i-1])*h)
        
    return t,y
```
