# ADAM BASHFORD METHOD

```python
def y1(t,y):
    return ((t**2-y**2)* np.sin(y))

def y2(t,y):
    return (t*2+ np.exp(-1*t*y))

def y3(t,y):
    return ((y**2 + 2*t*y)/(3 + t**2))
```

```python
import numpy as np
import matplotlib.pyplot as plt


def ABM(h, t0, y0, f):
    t = np.arange(0.0, 2.0+0.001, h)
    y = np.zeros(t.size)
    y[0] = y0


    for i in range(0,t.size-1):    
        if i in range(0,3): 

            k1 = h * f(t[i],y[i])
            k2 = h * f(t[i] + (h/2.0), y[i] +(k1/2.0))
            k3 = h * f(t[i] + (h/2.0), y[i] + (k2/2.0))
            k4 = h * f(t[i] + h, y[i] + k3)

            y[i + 1] = y[i] + (k1 + 2.0*k2 + 2.0*k3 + k4)/6.0
        else:       
            y[i + 1] = y[i] + h*(55.0 * f(t[i],y[i]) - 59.0 * f(t[i-1],y[i-1]) + 37.0 * f(t[i-2],y[i-2]) - 9.0 * f(t[i-3],y[i-3]))/24.0

    return t,y
```

```python
T1,Y1 = ABM(0.05, 0.0, -1.0, y1)
T2,Y2 = ABM(0.1, 0.0, -1.0, y1)
plt.plot(T1, Y1, 'r--', label='h=0.05')
plt.plot(T2, Y2, 'b--', label='h=0.1')
plt.legend(loc='best')
plt.show()


```
